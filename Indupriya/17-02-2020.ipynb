```python

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'CSE', 'ECE', 'Dogs', 'Logs','cats','puppies',"a","b"
sizes = [10,10,10,10,10,10,10,10]
explode = (0, 0, 0, 0, 0,0 ,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, #autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()

```

## Lambda

- A simple one line function
- do not use def or return keywords 
- They are implicit

```python
# for ex

def double(x):
    return x**2
```

```python
double(9)
```

```python
# using lambda
# - Here x is a parameter and return is an expression

a=lambda x:x**2
print(a(3))
```

```python
# using two parameters

def add(x,y):
    return x+y
add(9,8)
```

```python
# using two parameters in lambda
b=lambda x,y:x+y
print(b(2,3))
```

```python
### map

# - Apply same function to each  element of a sequence

# - Return modified list
# square list


def square(list1):
    list2=[]
    for num in list1:
        list2.append(num**2)
    return list2
```

```python
square([3,4,6,7,8])
```

```python
# using map
n=[1,2,3,4,5]
my_map=list(map(lambda x:x*2*7,n))
print(my_map)
```

```python
# using without map functio

square([9,8,7,6])
```

```python
## filter

# -Items are out of sequence
# Return filter list
n=[9,7,6,4,2,4,5,0,1,1,-2,-5,-7,-4]
my_filter=list(filter(lambda x:x>2,n))
print(my_filter)
```
